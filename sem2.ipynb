```python
titanic = pd.read_csv('https://raw.githubusercontent.com/iad34/seminars/master/materials/data_sem1.csv', sep=';')
titanic
```

```python
titanic[titanic.Sex == 'male']['Age'].mean()
```

```python
titanic[titanic.Sex == 'female']['Age'].mean()
```

```python
titanic.groupby(['Sex'])['Age'].mean()
```

```python
titanic.groupby('Survived').mean()
```

```python
titanic['Surname']=titanic['Name'].map(lambda x: x.split(';')[0])
```

```python
titanic
```

```python
surnames = titanic.groupby('Surname').apply(lambda x: x.shape[0])
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2*np.pi, num = 100)
y = np.sin(x)
z = np.cos(x)
```

```python
plt.plot(x, y)
plt.plot(x, z)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Functions')
plt.legend(['sin','cos'])
#можно сохранять как PDF
```
